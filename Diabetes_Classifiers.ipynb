```python
X =df.loc[:,['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']]
Y = df.loc[:,'Outcome']
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_test_split?
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model1 = LogisticRegression()
model1.fit(X_train,y_train)
print("Train Accuracy: ", model1.score(X_train,y_train))
print("Test Accuracy: ", model1.score(X_test,y_test))
```

```python
predict1 = model1.predict(X_test)
```

```python
# Importing evaluation Metrics
from sklearn.metrics import classification_report, f1_score, confusion_matrix, log_loss,jaccard_similarity_score, accuracy_score
```

```python
print("Classification Report")
print(classification_report(y_test, predict1))
print("Confusion Matrix")
print(confusion_matrix(y_test, predict1))
print("F1- Score")
print(f1_score(y_test, predict1))
print("Log Loss")
print(log_loss(y_test, predict1))
print("Jaccard Similarity Index")
print(jaccard_similarity_score(y_test, predict1))
print("Accuracy Score")
print(accuracy_score(y_test, predict1))
```
